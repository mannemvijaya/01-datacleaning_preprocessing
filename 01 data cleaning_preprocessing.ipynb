```python
#there is missing values so filling with 0 without removes for other data
df = pd.read_csv('train.csv')
df['Postal Code'] = df['Postal Code'].fillna(0)
```

```python
df.isnull().sum()
```

```python
#find duplicates
df.duplicated().sum() 
```

```python
# to make colum names in same format
df.columns = df.columns.str.lower().str.replace(" ", "_").str.strip()
```

```python
df.head()
```

```python
#converting string date type into actual date format
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
df['ship_date'] = pd.to_datetime(df['ship_date'], dayfirst=True)
```

```python
df.dtypes
```

```python
#converting float dtype into int for postal code
df['postal_code'] = df['postal_code'].astype(int)
```

```python
#saving cleaned data set
df.to_csv("cleaned_dataset.csv", index=False)
```
